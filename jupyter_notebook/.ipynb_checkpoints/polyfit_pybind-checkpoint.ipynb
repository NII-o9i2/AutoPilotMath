```python
points_list = pybind_test.gen_random_points(0.1,0,1,0,0,1,0.2,0.05,0.001,15)
eigen_coef = pybind_test.poly_fit_eigen_python(points_list)
osqp_coef = pybind_test.poly_fit_osqp_python(points_list)
spline_points = pybind_test.poly_fit_spline_python(points_list)
```

```python
x = []
y = []
y_eigen = []
y_osqp_c = []

for point in points_list:
    x.append(point.x)
    y.append(point.y)
    y_eigen.append(eigen_coef[0] * point.x **3 + eigen_coef[1] * point.x **2 + eigen_coef[2] * point.x  + eigen_coef[3] )
    y_osqp_c.append(osqp_coef[3] * point.x **3 + osqp_coef[2] * point.x **2 + osqp_coef[1] * point.x  + osqp_coef[0] )
    
x_spline = []
y_spline = []

for point in spline_points:
    x_spline.append(point.x)
    y_spline.append(point.y)
```

```python
import osqp
import numpy as np
from scipy import sparse
x_i_pow = []

point_num = len(points_list)
# print(x[0:point_num])
# print(y[0:point_num])
order = 3
for x_tmp in x:
    x_tmp_list = [1, x_tmp, x_tmp *x_tmp , x_tmp *x_tmp*x_tmp]
    x_i_pow.append(x_tmp_list)


p = [[0 for j in range(order + 1)] for i in range(order + 1)]

for i in range(order + 1):
    for j in range(order + 1):
        sum_tmp = 0.0
        for k in range(point_num):
            sum_tmp = sum_tmp + x_i_pow[k][i] * x_i_pow[k][j]
        p[i][j] = 2*sum_tmp
# print("p is",p)
q = []

for i in range(order+1):
    sum_tmp = 0.0
    for k in range(point_num):
        sum_tmp = sum_tmp - 2* y[k] * x_i_pow[k][i]
    q.append(sum_tmp)
# print('q is ', q)
        
a = [[0,0,0,1]]

# Define problem data
P = sparse.csc_matrix(p)
# print(P)
q = np.array(q)
A = sparse.csc_matrix(a)
l = np.array([-100000])
u = np.array([1000000])

# # Create an OSQP object
prob = osqp.OSQP()

# # # # Setup workspace and change alpha parameter
prob.setup(P, q, A, l, u, alpha=1.0)

# # # # # Solve problem
res = prob.solve()
count = 1
print(res.x)
y_osqp = []
for i in range(point_num):
    y_osqp.append(res.x[0] + res.x[1] * x[i] + res.x[2] * x[i]**2  + res.x[3] * x[i] **3)
#     print("y fit is ",res.x[0] + res.x[1] * x[i] + res.x[2] * x[i]**2  + res.x[3] * x[i] **3 )
#     print("y raw is ", y[i])
```
