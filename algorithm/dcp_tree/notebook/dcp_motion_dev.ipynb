```python
import sys
import numpy as np
import os
import copy
current_directory = os.getcwd()
relative_path1 = '../../build/env_simulator'
relative_path2 = '../../build/dcp_tree'
file_path1 = os.path.join(current_directory, relative_path1)
file_path2 = os.path.join(current_directory, relative_path2)
sys.path.append(file_path1)
sys.path.append(file_path2)
import pybind_dcp_tree
import pybind_env_simulator
import bokeh
import math
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.layouts import column, row
from bokeh.io import output_notebook,output_file
from bokeh.palettes import Magma, Viridis, Category20
from bokeh.models import Arrow, VeeHead
from bokeh.models import WheelZoomTool
from bokeh.models import HoverTool
from bokeh.models import Ellipse
```

```python
def plot_dcp_motion(objs,traj_new, res_pts, open_points):
    p = figure(width=900, height=400, match_aspect=True)

    step_vh = 2.0

    vehicle_length = 4.9
    vehicle_width = 1.9

    vh = VeeHead(size=5, fill_color=Viridis[3][0])

    # result traj
    x_new = []
    y_new = []
    x_new_collision = []
    y_new_collision = []
    x_open = []
    y_open = []
    theta_new = []
    theta_new_collision = []
    theta_open = []
    x_text = []
    y_text = []
    text_l = []
    
    x_res = []
    y_res = []
    theta_res = []
    index = 0
    for pt in traj_new:
        if pt["collision"]:
            x_new_collision.append(pt['x'])
            y_new_collision.append(pt['y'])
            theta_new_collision.append(pt['theta'])
        else:
            x_new.append(pt['x'])
            y_new.append(pt['y'])
            theta_new.append(pt['theta'])
        
        # draw arrow
        next_pt_x = pt['x'] + math.cos(pt['theta']) * step_vh
        next_pt_y = pt['y'] + math.sin(pt['theta']) * step_vh
        # veh_vec_len_front_x = 0.5 * vehicle_length * math.cos(pt['theta']) 
        # veh_vec_len_front_y = 0.5 * vehicle_length * math.sin(pt['theta']) 
        # veh_vec_wid_left_x = 0.5 * vehicle_width * math.cos(pt['theta'] + 0.5 * math.pi)
        # veh_vec_wid_left_y = 0.5 * vehicle_width * math.sin(pt['theta'] + 0.5 * math.pi) 
        # r_l = [[1,1],[1,-1],[-1,-1],[-1,1]]
        ploy_gon_x = []
        ploy_gon_y = []
        for ploy in pt['polygon']:
            ploy_gon_x.append(ploy.x)
            ploy_gon_y.append(ploy.y)
        x_text.append(pt['x'])
        y_text.append(pt['y'])
        if pt['collision']:
            p.patch(ploy_gon_x, ploy_gon_y, alpha=0.2, line_width=2, color = Magma[6][4],legend_label="collision traj") 
        else:
            p.patch(ploy_gon_x, ploy_gon_y, alpha=0.2, line_width=2, color = Magma[6][3],legend_label="traj") 
        p.add_layout(Arrow(end=vh, x_start=pt['x'], y_start=pt['y'], x_end=next_pt_x, y_end=next_pt_y))
        index = index + 1
        
    for pt in open_points:
        x_open.append(pt['x'])
        y_open.append(pt['y'])
        theta_open.append(pt['theta'])
        
        # draw arrow
        next_pt_x = pt['x'] + math.cos(pt['theta']) * step_vh
        next_pt_y = pt['y'] + math.sin(pt['theta']) * step_vh
        # veh_vec_len_front_x = 0.5 * vehicle_length * math.cos(pt['theta']) 
        # veh_vec_len_front_y = 0.5 * vehicle_length * math.sin(pt['theta']) 
        # veh_vec_wid_left_x = 0.5 * vehicle_width * math.cos(pt['theta'] + 0.5 * math.pi)
        # veh_vec_wid_left_y = 0.5 * vehicle_width * math.sin(pt['theta'] + 0.5 * math.pi) 
        # r_l = [[1,1],[1,-1],[-1,-1],[-1,1]]
        ploy_gon_x = []
        ploy_gon_y = []
        for ploy in pt['polygon']:
            ploy_gon_x.append(ploy.x)
            ploy_gon_y.append(ploy.y)
        x_text.append(pt['x'])
        y_text.append(pt['y'])
        p.patch(ploy_gon_x, ploy_gon_y, alpha=0.2, line_width=2, color = Magma[6][3],legend_label="traj") 
        p.add_layout(Arrow(end=vh, x_start=pt['x'], y_start=pt['y'], x_end=next_pt_x, y_end=next_pt_y))
        
    for pt in res_pts:
        x_res.append(pt['x'])
        y_res.append(pt['y'])
        theta_res.append(pt['theta'])
        
        # draw arrow
        next_pt_x = pt['x'] + math.cos(pt['theta']) * step_vh
        next_pt_y = pt['y'] + math.sin(pt['theta']) * step_vh
        # veh_vec_len_front_x = 0.5 * vehicle_length * math.cos(pt['theta']) 
        # veh_vec_len_front_y = 0.5 * vehicle_length * math.sin(pt['theta']) 
        # veh_vec_wid_left_x = 0.5 * vehicle_width * math.cos(pt['theta'] + 0.5 * math.pi)
        # veh_vec_wid_left_y = 0.5 * vehicle_width * math.sin(pt['theta'] + 0.5 * math.pi) 
        # r_l = [[1,1],[1,-1],[-1,-1],[-1,1]]
        ploy_gon_x = []
        ploy_gon_y = []
        for ploy in pt['polygon']:
            ploy_gon_x.append(ploy.x)
            ploy_gon_y.append(ploy.y)
        x_text.append(pt['x'])
        y_text.append(pt['y'])
        p.patch(ploy_gon_x, ploy_gon_y, alpha=0.6, line_width=2, color = Magma[6][4],legend_label="res") 
        p.add_layout(Arrow(end=vh, x_start=pt['x'], y_start=pt['y'], x_end=next_pt_x, y_end=next_pt_y))
            
    
    
    source = ColumnDataSource(data=dict(
        x=x_new,
        y=y_new,
        theta=theta_new,
    ))
    source_exp = ColumnDataSource(data=dict(
        x=x_new_collision,
        y=y_new_collision,
        theta=theta_new_collision,
    ))
    scatter_new = p.scatter('x', 'y', source=source, line_width=2, size = 8 ,color = Magma[6][3], legend_label="traj_new")
    scatter_new_exp = p.diamond('x', 'y', source=source_exp, line_width=2, size = 10 ,color = Magma[6][4], legend_label="traj_new_exp")
    # p.text(x_text,y_text, text=text_l,text_font_size = "7pt")
    hover_tool = HoverTool(renderers=[scatter_new], tooltips=[
                                                            ("x", "@x"),
                                                            ("y", "@y"),
                                                            ("theta", "@theta"),
                                                            ], mode='mouse')
    p.add_tools(hover_tool)
    # 
    # # objs
    for obj in objs:
        text_l = []
        obj_a = obj['ellipse_a']
        obj_b = obj['ellipse_b']
        obj_theta = obj['theta']
        obj_x = obj['x']
        obj_y = obj['y']
        p.ellipse(x=obj_x, y=obj_y, width=obj_a*2, height=obj_b*2, angle=obj_theta, fill_alpha = 0.5, line_alpha=0.5, fill_color="green", line_color="white")
        # 
        # p.scatter(pts_tmp_x, pts_tmp_y, line_width=2, color = Viridis[6][a], size = 10,legend_label="obj")
        # p.text(pts_tmp_x, pts_tmp_y, text=text_l,text_font_size = "7pt")
        # a = a +1
    # 
    p.toolbar.active_scroll = p.select_one(WheelZoomTool)
    p.legend.click_policy = 'hide'
    output_notebook()

    show(p)

```

```python
def process_init_state(init_theta, init_omega, init_vel):
    step_size = 9
    delta_t = 1
    state = pybind_dcp_tree.DCPMotionTreePoint()
    next_state = pybind_dcp_tree.DCPMotionTreePoint()
    state.index = 1
    state.theta = init_theta
    state.omega = init_omega
    state.speed_limit = 40
    state.v = init_vel
    state.point.x = 0.0
    state.point.y = 0.0
    state.length = 4.9
    state.width = 2.0
    
    init_points = []
    init_points_py = []
    
    motion_tree = pybind_dcp_tree.DCPMotionTree()
    
    for i in range(step_size):
        state.update_polygon()
        # state_new = pybind_dcp_tree.DCPMotionTreePoint()
        state_new = pybind_dcp_tree.DCPMotionTreePoint(state)
        init_points.append(state_new)
        point_py = {}
        point_py['x'] = state_new.point.x
        point_py['y'] = state_new.point.y
        point_py['theta'] = state_new.theta
        point_py['polygon'] = [state_new.polygon[0], state_new.polygon[1], state_new.polygon[2],state_new.polygon[3]]
        init_points_py.append(point_py)
        
        pybind_dcp_tree.DCPMotionTree.step_a_omega_dcp(state,next_state,delta_t)
        state = next_state
    
    obj1 = pybind_dcp_tree.DCPObjectInfo()
    obj1.center_point.x = 25.0
    obj1.center_point.y = 0.0
    obj1.theta = 0.523
    obj1.ellipse_a = 3.0
    obj1.ellipse_b = 1.8
    
    obj2 = pybind_dcp_tree.DCPObjectInfo()
    obj2.center_point.x = 50.0
    obj2.center_point.y = 5.0
    obj2.theta = 0.2
    obj2.ellipse_a = 3.0
    obj2.ellipse_b = 1.8
    
    obj3 = pybind_dcp_tree.DCPObjectInfo()
    obj3.center_point.x = 40.0
    obj3.center_point.y = -6.0
    obj3.theta = -0.1
    obj3.ellipse_a = 3.2
    obj3.ellipse_b = 2.0
    
    obj1_py = {
        'x': obj1.center_point.x,
        'y': obj1.center_point.y,
        'theta': obj1.theta,
        'ellipse_a': obj1.ellipse_a,
        'ellipse_b': obj1.ellipse_b
    }
    obj2_py = {
        'x': obj2.center_point.x,
        'y': obj2.center_point.y,
        'theta': obj2.theta,
        'ellipse_a': obj2.ellipse_a,
        'ellipse_b': obj2.ellipse_b
    }
    obj3_py = {
        'x': obj3.center_point.x,
        'y': obj3.center_point.y,
        'theta': obj3.theta,
        'ellipse_a': obj3.ellipse_a,
        'ellipse_b': obj3.ellipse_b
    }
    
    obj1_py_list = [obj1_py,obj2_py,obj3_py]

    obj_list = [obj1,obj2,obj3]
    
    # for obj in obj_list:
    for i in range(len(init_points) -1):
        crash = False
        for obj in obj_list:
            opt = pybind_dcp_tree.DCPMotionTree.check_collision(init_points[i],init_points[i+1],obj)
            if opt:
                crash = True
                break
        init_points_py[i]['collision'] = crash
    init_points_py[len(init_points)-1]['collision'] = init_points_py[len(init_points) - 2]['collision']
                
            
    
    param = pybind_dcp_tree.DCPMotionTreeParameter()
    motion_tree.init(init_points,obj_list,delta_t,param)
    motion_tree.process()
    open_points = motion_tree.get_open_points()
    res_points = motion_tree.get_result_points()

    open_points_py = []
    for point in open_points:
        point_py = {}
        point_py['x'] = point.point.x
        point_py['y'] = point.point.y
        point_py['theta'] = point.theta
        point_py['polygon'] = [point.polygon[0], point.polygon[1], point.polygon[2],point.polygon[3]]
        open_points_py.append(point_py)
        
    res_points_py = []
    for point in res_points:
        point_py = {}
        point_py['x'] = point.point.x
        point_py['y'] = point.point.y
        point_py['theta'] = point.theta
        point_py['polygon'] = [point.polygon[0], point.polygon[1], point.polygon[2],point.polygon[3]]
        res_points_py.append(point_py)
    plot_dcp_motion(obj1_py_list, init_points_py,res_points_py, open_points_py)
    return motion_tree
```
