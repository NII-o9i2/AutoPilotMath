## freespace case 1

```python
env = pybind_env_simulator.EnvSimulator()
path = str(current_directory)+str("/data/freespace_case_2.json")
env.update_case_data(path)
pts = env.get_all_center_points()
obs = env.get_all_obstacle()
road_edge_pts = env.get_all_road_edge_points()
freespace_roi_pts = env.get_freespace_points_by_roi()
ego_car = env.get_ego_car()
ellipse_param = env.get_freespace_ellipse_param()

planning_point = pybind_env_simulator.PlanningPoint()
planning_point.position.x = 0
planning_point.position.y = 2
planning_point.theta = 0
planning_point.velocity = 10
planning_point.curva = 0
planning_point.acceleration = 0


right_point_by_ttc = env.get_freespace_right_points_by_ttc(7,planning_point)
print('right_point_by_ttc size:',len(right_point_by_ttc))

ego_prediction_position = env.get_predict_ego_position_at_time(7,planning_point)

right_border_pts = env.get_right_border_use_ellipse_model(planning_point)


py_env_sim.plot_freespace(pts,obs,road_edge_pts,freespace_roi_pts,right_point_by_ttc,ego_prediction_position,right_border_pts,planning_point,ellipse_param)
```
